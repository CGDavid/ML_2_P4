# Handout #4: Unsupervised Learning (hierarchical clustering)

```python
import numpy as np
from sklearn import cluster
import matplotlib.pyplot as plt
```

```python
group = '03' # assuming group 1
ds = 5 # assuming problems 5
data = np.loadtxt(f'ds{group}{str(ds)}.txt')
X = data[:, 0:2]
y = data[:, 2]
```

## Pintado de datos con etiquetas originales 

```python
def draw_dataset(data):
    plt.subplot(111,aspect=1)
    w_0 = data[data[:,2] == 0]
    w_1 = data[data[:,2] == 1]
    w_2 = data[data[:,2] == 2]
    plt.plot(w_0[:,0], w_0[:,1], 'rx')
    plt.plot(w_1[:,0], w_1[:,1], 'bx');
    plt.plot(w_2[:,0], w_2[:,1], 'gx');
    
draw_dataset(data)
```

## Soluciones

```python
def draw_clusters(X, labels):
    distinct_labels = list(set(labels))
    
    for label in distinct_labels:
        plt.scatter(X[labels == label, 0], X[labels == label, 1], marker='x', color=np.random.rand(3,))
```

```python
average = cluster.AgglomerativeClustering(
    n_clusters=3, linkage="average"
)

average.fit(X)

average_labels = average.labels_

draw_clusters(X, average_labels)
```

```python
complete = cluster.AgglomerativeClustering(
    n_clusters=3, linkage="complete"
)

complete.fit(X)

complete_labels = complete.labels_

draw_clusters(X, complete_labels)
```

```python
ward = cluster.AgglomerativeClustering(
    n_clusters=3, linkage="ward"
)

ward.fit(X)

ward_labels = ward.labels_

draw_clusters(X, ward_labels)
```

```python
clustering_algorithms = (
    ("Average Linkage", average),
    ("Complete Linkage", complete),
    ("Ward Linkage", ward),
)

plot_num = 0

for name, model in clustering_algorithms:
    plt.subplot(1, len(clustering_algorithms), plot_num)

    model.fit(X)
    ward_labels = ward.labels_
    draw_clusters(X, ward_labels)
    
    plot_num = plot_num + 1
    
```
